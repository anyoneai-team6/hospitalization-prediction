```python
def seleccionar_elementos(lista,x):
    """
    Select elements from a list that start with a specific character or string.

    Args:
        lista (list): The input list of elements.
        x (str): The character or string to check for at the beginning of each element.

    Returns:
        list: A list of selected elements.

    """
    seleccionados = [elemento for elemento in lista if str(elemento).startswith(x)]
    return seleccionados


def cambiar_letras(lista, x):
    """
    Modify elements in a list by replacing the first character or characters with a specific character or string.

    Args:
        lista (list): The input list of elements.
        x (str): The character or string to replace the first character(s) in each element.

    Returns:
        list: A list of modified elements.

    """
    modificados = [x + elemento[2:] for elemento in lista]
    return modificados


def show_data(df):
    """
    Print the shape and first few rows of a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: The same DataFrame as the input.

    """
    print(df.shape)
    return df.head()


def check_columns_existence(column_list:list, dataframe:pd.DataFrame):
    """
    Check the existence of columns in a DataFrame.

    Args:
        column_list (list): List of column names to check.
        dataframe (pd.DataFrame): The input DataFrame.

    Returns:
        tuple: A tuple containing two lists - existing columns and non-existing columns.

    """
    existing_columns = []
    non_existing_columns = []

    for column in column_list:
        if column in dataframe.columns:
            existing_columns.append(column)
        else:
            non_existing_columns.append(column)

    return existing_columns, non_existing_columns


def transform_dataframes(df_r, df, encode):
    """
    Transform DataFrames by modifying column names and combining data.

    Args:
        df_r (pd.DataFrame): The first input DataFrame.
        df (pd.DataFrame): The second input DataFrame.
        encode (str): The character or string used for column name transformation.

    Returns:
        pd.DataFrame: The transformed DataFrame.

    """
    col = cambiar_letras(df_r.columns, encode)
    exist, no_exist = check_columns_existence(col, df)
    df_r_new = pd.DataFrame()
    df_r_new[cambiar_letras(exist, 'pa')] = df[exist]
    df_r_new[cambiar_letras(no_exist, 'pa')] = df_r[cambiar_letras(no_exist, 'pa')]
    
    return df_r_new
```
